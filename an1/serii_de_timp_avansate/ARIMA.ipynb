# Time series

\begin{array}{l}{\text { Denote the mean function of }\left\{X_{t}\right\} \text { as }} \\ {\qquad \mu_{X}(t)=\mathrm{E}\left(X_{t}\right)} \\ {\text { provided it exists. And the autocovariance function of }\left\{X_{t}\right\} \text { is }} \\ {\qquad \gamma_{X}(s, t)=\operatorname{Cov}\left(X_{s}, X_{t}\right)=\mathrm{E}\left[\left\{X_{s}-\mu_{X}(s)\right\}\left\{X_{t}-\mu_{X}(t)\right\}\right]}\end{array}

\begin{array}{l}{\text { The time series }\left\{X_{t}\right\} \text { is (weakly) stationary if }} \\ {\begin{array}{l}{\text { 1. } \mu_{X}(t) \text { is independent of } t} \\ {\text { 2. } \gamma_{X}(t+h, h) \text { is independent of } t \text { for each } h .} \\ {\text { We say }\left\{X_{t}\right\} \text { is strictly (or strongly) stationary if }} \\ {\qquad\left(X_{t_{1}}, \ldots, X_{t_{k}}\right) \text { and }\left(X_{t_{1}+h}, \ldots, X_{t_{k}+h}\right) \text { have the same joint distributions }} \\ {\text { for all } k=1,2, \ldots, h=0, \pm 1, \pm 2, \ldots, \text { and time points } t_{1}, \ldots, t_{k} . \text { This is a very strong condition. }}\end{array}}\end{array}

\begin{array}{l}{\text { Let }\left\{X_{t}\right\} \text { be a stationary time series. Its mean is } \mu_{X}=\mu_{X}(t) . \text { Its autocovariance function }} \\ {\begin{aligned}(\mathrm{ACVF}) \text { of }\left\{X_{t}\right\} \text { at lag } h \text { is } & \\ \text { Its autocorrelation function }(\mathrm{ACF}) & \text { of }\left\{X_{t}\right\} \text { at } \operatorname{lag} h \text { is } \\ \rho_{X}(h)=\frac{\gamma_{X}(h)}{\gamma_{X}(0)} &=\operatorname{Corr}\left(X_{t+h}, X_{t}\right) \end{aligned}}\end{array}

\begin{array}{cl}{\text { Time series: }} & {\text { Sample ACF }} \\ {\text { White noise }} & {\text { Zero for }|h|>0} \\ {\text { Trend }} & {\text { Slow decay }} \\ {\text { Periodic }} & {\text { Periodic }} \\ {\operatorname{MA}(q)} & {\text { Zero for }|h|>q} \\ {\operatorname{AR}(p)} & {\text { Decays to zero exponentially}} \end{array}

\begin{array}{l}{\text { The typical procedure of time series modeling can be described as }} \\ {\text { 1. Plot the time series (look for trends, seasonal components, step changes, outliers). }} \\ {\text { 2. Transform data so that residuals are stationary. }} \\ {\text { (a) Estimate and subtract } m_{t}, s_{t}} \\ {\text { (b) Differencing }} \\ {\text { (c) Nonlinear transformations }(\log , \sqrt{\cdot})} \\ {\text { 3. Fit model to residuals. }}\end{array}

\begin{array}{l}{\text { Define the lag-1 difference operator: (think 'first derivative') }} \\ {\qquad \nabla X_{t}=\frac{X_{t}-X_{t-1}}{t-(t-1)}=X_{t}-X_{t-1}=(1-B) X_{t}} \\ {\text { where } B \text { is the backshift operator, } B X_{t}=X_{t-1}} \\ {\text { Define the lag-s difference operator, }} \\ {\qquad \nabla_{s} X_{t}=X_{t}-X_{t-s}=\left(1-B^{s}\right) X_{t}} \\ {\text { where } B^{s} \text { is the backshift operator applied } s \text { times, } B^{s} X_{t}=B\left(B^{s-1} X_{t}\right) \text { and } B^{1} X_{t}=B X_{t}.}\end{array}.

## ARMA (p,q) Models


 An ARMA(p, q) process $X_{t}$  is a stationary process that satisfies:
$X_{t}-\phi_{1} X_{t-1}-\cdots-\phi_{p} X_{t-p}=\epsilon_{t}+\theta_{1} 
\epsilon_{t-1}+\cdots+\theta_{q} \epsilon_{t-q}$.

This can also be written as:
$\phi(B) X_{t}=\theta(B) \epsilon_{t}$, where
$\phi(B)=1-\phi_{1} B-\cdots-\phi_{p} B^{p}$ and $\theta(B)=1+\theta_{1} B+\cdots+\theta_{q} B^{q}$ and 
$\epsilon_{t}\sim \mathrm{WN}(0, \sigma^{2})$.

\begin{array}{l}{\left.\bullet\left\{X_{t}\right\} \text { is said to be causal (or more specifically to be a causal function of }\left\{\epsilon_{t}\right\}\right) \text { if there exits a }} \\ {\text { sequence of constants }\left\{\psi_{j}\right\} \text { such that } \sum_{j=0}^{\infty}\left|\psi_{j}\right|<\infty \text { and }} \\ {\qquad X_{t}=\sum_{j=0}^{\infty} \psi_{j} \epsilon_{t-j}, \quad t=0, \pm 1, \ldots} \\ {\left.\bullet\left\{X_{t}\right\} \text { is said to be invertible (or more specifically to be an invertible function of }\left\{\epsilon_{t}\right\}\right) \text { if there }} \\ {\text { exits a sequence of constants }\left\{\pi_{j}\right\} \text { such that } \sum_{j=0}^{\infty}\left|\pi_{j}\right|<\infty \text { and }} \\ {\epsilon_{t}=\sum_{j=0}^{\infty} \pi_{j} X_{t-j}, \quad t=0, \pm 1, \ldots}\end{array}

\begin{array}{l}{\left.\text { Theorem: Let }\left\{X_{t}\right\} \text { be an ARMA(p, } q\right) \text { process. Then }\left\{X_{t}\right\} \text { is causal if and only if }} \\ {\qquad \phi(z) \neq 0 \text { for all }|z| \leq 1.}\end{array}

# ARIMA (p,d,q) Models

If d  is a non-negative integer, then ${X_{t}}$ is said to be  an ARIMA(p, d, q)  process if ${X_{t}}$ is 
1. intrinsically stationary of order  d.

2. $\nabla^{d} X_{t}$ is a causal ARMA(p, q) process. 

## How to find the order of differencing (d) in ARIMA model

```python
import numpy as np, pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

# Import data
raw_data = open('pibt.txt','r').read().split('\t')
df = pd.DataFrame([float(item) for item in raw_data[1:]], columns = ['value'],)

# Original Series
fig, axes = plt.subplots(5, 2, sharex=True)
axes[0, 0].plot(df.value); axes[0, 0].set_title('Original Series')
plot_acf(df.value, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(df.value.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.value.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(df.value.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df.value.diff().diff().dropna(), ax=axes[2, 1])

# Log stationarity

axes[3, 0].plot(np.log(df.value.dropna())); axes[3, 0].set_title('Log Stationarity')
plot_acf(np.log(df.value.dropna()), ax=axes[3, 1])

# Log stationarity with 1st order diff

axes[4, 0].plot(np.log(df.value.dropna()).diff()); axes[4, 0].set_title('Log Stationarity with first order diff')
plot_acf(np.log(df.value.dropna()).diff(), ax=axes[4, 1])

plt.show()
```

## Stationarity tests

\begin{array}{l}{\text { The Dickey-Fuller test is testing if } \phi=0 \text { in this model of the data: }} \\ {\qquad y_{t}=\alpha+\beta t+\phi y_{t-1}+e_{t}} \\ {\text { which is written as }} \\ {\qquad \Delta y_{t}=y_{t-1}=\alpha+\beta t+\gamma y_{t-1}+e_{t}} \\ {\text { where } y_{t} \text { is your data. It is written this way so we can do a linear regression of } \Delta y_{t} \text { against } t \text { and } y_{t-1} \text { and }} \\ {\text { test if } \gamma \text { is different from } 0 . \text { If } \gamma=0 \text { , then we have a random walk process. If not and }-1<1+\gamma<1 \text { , }} \\ {\text { then we have a stationary process. }}\end{array}

\begin{array}{l}{\text { The Augmented Dickey-Fuller test allows for higher-order autoregressive processes by including } \Delta y_{t-p}} \\ {\text { in the model. But our test is still if } \gamma=0} \\ {\qquad \Delta y_{t}=\alpha+\beta t+\gamma y_{t-1}+\delta_{1} \Delta y_{t-1}+\delta_{2} \Delta y_{t-2}+\ldots} \\ {\text { The null hypothesis for both tests is that the data are non-stationary. We want to REJECT the null }} \\ {\text { hypothesis for this test, so we want a p-value of less that } 0.05 \text { (or smaller). }}\end{array}

```python
## Adf Test
from pmdarima.arima.utils import ndiffs
y = np.log(df.value.dropna()).diff().dropna()
from statsmodels.tsa.stattools import adfuller
from numpy import log
result = adfuller(y.dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print(ndiffs(y, test='adf'))  # 2
```

## Accuracy Metrics for Time Series Forecast
    Mean Absolute Percentage Error (MAPE)
    Mean Error (ME)
    Mean Absolute Error (MAE)
    Mean Percentage Error (MPE)
    Root Mean Squared Error (RMSE)
    Lag 1 Autocorrelation of Error (ACF1)
    Correlation between the Actual and the Forecast (corr)
    Min-Max Error (minmax)
Typically, if you are comparing forecasts of two different series, the MAPE, Correlation and Min-Max Error can be used.

Why not use the other metrics?

Because only the above three are percentage errors that vary between 0 and 1. That way, you can judge how good is the forecast irrespective of the scale of the series.

The other error metrics are quantities. That implies, an RMSE of 100 for a series whose mean is in 1000s is better than an RMSE of 5 for series in 10s. So, you cant really use them to compare the forecasts of two different scaled time series.

```python
# Accuracy metrics
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    acf1 = acf(fc-test)[1]                      # ACF1
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 'acf1':acf1, 
            'corr':corr, 'minmax':minmax})

forecast_accuracy(fc, test.values)
```

## How to do Auto Arima Forecast in Python

```python
from statsmodels.tsa.arima_model import ARIMA
import pmdarima as pm

model = pm.auto_arima(df.value, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
print(model.summary())
```

## How to interpret the residual plots in ARIMA model

```python
model.plot_diagnostics(figsize=(7,5))
plt.show()
```

Top left: The residual errors seem to fluctuate around a mean of zero and have a uniform variance.

Top Right: The density plot suggest normal distribution with mean zero.

Bottom left: All the dots should fall perfectly in line with the red line. Any significant deviations would imply the distribution is skewed.

Bottom Right: The Correlogram, aka, ACF plot shows the residual errors are not autocorrelated. Any autocorrelation would imply that there is some pattern in the residual errors which are not explained in the model. So you will need to look for more Xs (predictors) to the model.

```python
# Forecast
n_periods = 24
fc, confint = model.predict(n_periods=n_periods, return_conf_int=True)
index_of_fc = np.arange(len(df.value), len(df.value)+n_periods)

# make series for plotting purpose
fc_series = pd.Series(fc, index=index_of_fc)
lower_series = pd.Series(confint[:, 0], index=index_of_fc)
upper_series = pd.Series(confint[:, 1], index=index_of_fc)
# Plot
plt.plot(df.value)
plt.plot(fc_series, color='darkgreen')
plt.fill_between(lower_series.index, 
                 lower_series, 
                 upper_series, 
                 color='k', alpha=.15)

plt.title("Final Forecast of GDP")
plt.show()
```

## How to automatically build SARIMA model in python

If your time series has defined seasonality, then, go for SARIMA which uses seasonal differencing.

Seasonal differencing is similar to regular differencing, but, instead of subtracting consecutive terms, you subtract the value from previous season.

So, the model will be represented as SARIMA(p,d,q)x(P,D,Q), where, P, D and Q are SAR, order of seasonal differencing and SMA terms respectively and 'x' is the frequency of the time series.

If your model has well defined seasonal patterns, then enforce D=1 for a given frequency x.

Heres some practical advice on building SARIMA model:

As a general rule, set the model parameters such that D never exceeds one. And the total differencing d + D never exceeds 2. Try to keep only either SAR or SMA terms if your model has seasonal components.

\begin{array}{l}{\text { The SARIMA(p, } d, q) \times(P, D, Q)_{s} \text { Process: If } d \text { and } D \text { are non-negative integers, then }\left\{X_{t}\right\} \text { is is }} \\ {\text { said to be a seasonal ARIMA(p, } d, q) \times(P, D, Q)_{s} \text { process with period } s \text { if the differenced process }} \\ {Y_{t}=(1-B)^{d}\left(1-B^{s}\right)^{D} X_{t} \text { is a causal ARMA process, }} \\ {\qquad \phi(B) \Phi\left(B^{s}\right) Y_{t}=\theta(B) \Theta\left(B^{s}\right) \epsilon_{t}, \quad \epsilon_{t} \sim \operatorname{WN}\left(0, \sigma^{2}\right)} \\ {\text { where } \phi(z)=1-\phi_{1} z-\cdots-\phi_{p} z^{p}, \Phi(z)=1-\Phi_{1} z-\cdots-\Phi_{p} z^{p}, \theta(z)=1+\theta_{1} z+\cdots+\theta_{q} z^{q} \text { and }} \\ {\Theta(z)=1+\Theta_{1} z+\cdots \Theta_{Q} z^{Q}.}\end{array}

```python
# Import
raw_data = open('pibt.txt','r').read().split('\t')
data = pd.DataFrame([float(item) for item in raw_data[1:]], columns = ['value'],)

# Plot
fig, axes = plt.subplots(2, 1, figsize=(10,5), dpi=100, sharex=True)

# Usual Differencing
axes[0].plot(data[:], label='Original Series')
axes[0].plot(data[:].diff(1), label='Usual Differencing')
axes[0].set_title('Usual Differencing')
axes[0].legend(loc='upper left', fontsize=10)


# Seasonal Differencing
axes[1].plot(data[:], label='Original Series')
axes[1].plot(data[:].diff(12), label='Seasonal Differencing', color='green')
axes[1].set_title('Seasonal Differencing')
plt.legend(loc='upper left', fontsize=10)
plt.suptitle('Quarterly GDP', fontsize=16)
plt.show()
```
